```python
from skimage.segmentation import slic
slic_img=slic(image,n_segments=30000,sigma=5)
```

```python
plt.imshow(slic_img,'gray')
plt.show()
from skimage.io import imsave
imsave("slic_img.jpg",slic_img)
```

```python
x=[]
y=[]
j=0
for (i, segVal) in enumerate(np.unique(slic_img)):
    mask = np.zeros(image.shape[:2], dtype = "uint8")
    print("inspecting segment",i)
    mask[slic_img == segVal] = 255
    seg=np.where(mask==255)
    #print("x coordiantes",seg[0])
    #print("y coordinates",seg[1])
    x.append(np.mean(seg[0]))
    y.append(np.mean(seg[1]))
    seg=np.zeros
x=np.array(x)
x=x.astype('int64')
y=np.array(y)
y=y.astype('int64')
```
