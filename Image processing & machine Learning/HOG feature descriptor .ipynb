```python
from skimage.io import imread
import numpy as np
import copy
from matplotlib import pyplot as plt
import cv2 
import os
def load_images(folder):
    images=[]
    for filename in os.listdir(folder):
        img=imread(os.path.join(folder,filename)) ## skimage.io.imread_collection(path)
        if img is not None:
            img=img*255
            img=img.astype('int64')
            images.append(img)
    return(images)

hog_images=load_images("D://Lokin//Data Analytics//TOMTOM//Images//HOG feature descriptor images")
```
